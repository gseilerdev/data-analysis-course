```python
import sqlite3
import pandas as pd
 
conexion = sqlite3.connect("vetcare.db")
 
# Qué tablas tiene la base de datos
tablas = pd.read_sql(
    "SELECT name FROM sqlite_master WHERE type='table'",
    conexion
)
print("Tablas:")
print(tablas)
 
# Estructura de cada tabla
for tabla in tablas["name"]:
    print(f"\nEstructura de '{tabla}':")
    estructura = pd.read_sql(f"PRAGMA table_info({tabla})", conexion)
    print(estructura[["name", "type"]])
 
conexion.close()
```

### Ejercicio 8 — Dos JOINs encadenados
 
Para ver el nombre del dueño junto con los datos de cada consulta médica, necesitas pasar por tres tablas: `consultas` → `mascotas` → `duenos`.
 
**Consigna:** Escribe una consulta que una las tres tablas y devuelva: nombre del dueño, nombre de la mascota, fecha de la consulta, motivo y costo. Ordená el resultado por fecha ascendente.

```python
import sqlite3
import pandas as pd
 
conexion = sqlite3.connect("vetcare.db")
 
df_historial = pd.read_sql(
    """
    SELECT duenos.nombre  AS dueno,
           mascotas.nombre AS mascota,
           consultas.fecha,
           consultas.motivo,
           consultas.costo
    FROM consultas
    JOIN mascotas ON consultas.mascota_id = mascotas.id
    JOIN duenos   ON mascotas.dueno_id    = duenos.id
    ORDER BY consultas.fecha ASC
    """,
    conexion
)
 
conexion.close()
 
print(df_historial)
```

### Ejercicio 9 — Investigación: alias en SQL
 
Cuando unes varias tablas, los nombres de las columnas pueden colisionar — dos tablas pueden tener una columna que se llama `id`, por ejemplo. Los alias resuelven ese problema.
 
**Consigna:** Investiga cómo usar `AS` en SQL para renombrar columnas y tablas en una consulta. Después reescribí la consulta del Ejercicio 8 usando alias para:
 
1. Renombrar `duenos.nombre` como `dueno` y `mascotas.nombre` como `mascota` en el resultado
2. Usar alias cortos para las tablas dentro de la consulta (por ejemplo, `c` para `consultas`, `m` para `mascotas`, `d` para `duenos`)

```python
import sqlite3
import pandas as pd
 
conexion = sqlite3.connect("vetcare.db")
 
df_historial = pd.read_sql(
    """
    SELECT d.nombre  AS dueno,
           m.nombre  AS mascota,
           c.fecha,
           c.motivo,
           c.costo
    FROM consultas AS c
    JOIN mascotas  AS m ON c.mascota_id = m.id
    JOIN duenos    AS d ON m.dueno_id   = d.id
    ORDER BY c.fecha ASC
    """,
    conexion
)
 
conexion.close()
 
print(list(df_historial.columns))
print(df_historial.head())
```

### Ejercicio 12 — Consultas condicionales con Python
 
En un sistema real, las consultas que hacés no siempre son las mismas — cambian según lo que el usuario pide o el análisis que estás haciendo. Necesitás construir consultas dinámicamente desde Python.
 
**Consigna:** Escribe una función `buscar_mascotas(especie=None, edad_maxima=None, ciudad=None)` que construya y ejecute una consulta SQL según los parámetros que reciba. Si un parámetro es `None`, no se incluye en el filtro. La función debe devolver un DataFrame.

```python
import sqlite3
import pandas as pd
 
def buscar_mascotas(especie=None, edad_maxima=None, ciudad=None):
    conexion = sqlite3.connect("vetcare.db")
 
    # Si se pide filtrar por ciudad, necesitamos JOIN con duenos
    necesita_join = ciudad is not None
 
    if necesita_join:
        base = "SELECT mascotas.* FROM mascotas JOIN duenos ON mascotas.dueno_id = duenos.id"
    else:
        base = "SELECT * FROM mascotas"
 
    condiciones = []
    parametros  = []
 
    if especie is not None:
        condiciones.append("mascotas.especie = ?" if necesita_join else "especie = ?")
        parametros.append(especie)
 
    if edad_maxima is not None:
        condiciones.append("mascotas.edad <= ?" if necesita_join else "edad <= ?")
        parametros.append(edad_maxima)
 
    if ciudad is not None:
        condiciones.append("duenos.ciudad = ?")
        parametros.append(ciudad)
 
    if condiciones:
        query = base + " WHERE " + " AND ".join(condiciones)
    else:
        query = base
 
    df = pd.read_sql(query, conexion, params=parametros)
    conexion.close()
    return df
 
 
print(buscar_mascotas(especie="perro"))
print(buscar_mascotas(edad_maxima=3))
print(buscar_mascotas(especie="gato", ciudad="Buenos Aires"))
print(buscar_mascotas())
```

### Ejercicio 13 — Reporte por veterinario
 
La clínica quiere saber cuánto facturó cada veterinario y cuántas consultas atendió. En la próxima clase vas a aprender `GROUP BY` y funciones de agregación en SQL — por ahora, hacelo con las herramientas que ya conocés.
 
**Consigna:** Trae todas las consultas con `pd.read_sql()` y calculá manualmente, usando Python puro (sin `.groupby()` ni `.mean()`), cuántas consultas atendió cada veterinario y cuánto facturó en total. Generá un reporte formateado.

```python
import sqlite3
import pandas as pd
 
conexion = sqlite3.connect("vetcare.db")
df = pd.read_sql("SELECT veterinario, costo FROM consultas", conexion)
conexion.close()
 
# Acumuladores por veterinario — sin groupby()
reporte = {}
 
for i in range(len(df)):
    vet   = df["veterinario"].iloc[i]
    costo = df["costo"].iloc[i]
 
    if vet not in reporte:
        reporte[vet] = {"consultas": 0, "total": 0.0}
 
    reporte[vet]["consultas"] += 1
    reporte[vet]["total"]     += costo
 
# Reporte formateado
print("REPORTE POR VETERINARIO")
print("─" * 40)
for vet, datos in sorted(reporte.items()):
    print(f"{vet}: {datos['consultas']} consultas — ${datos['total']:,.0f}")
```

> **Para pensar:** este ejercicio te muestra por qué `GROUP BY` existe. Anotá cuántas líneas necesitaste. En la próxima clase vas a resolver lo mismo en dos líneas con SQL.

### Ejercicio 14 — Múltiples consultas, una conexión
 
Abrir y cerrar una conexión para cada consulta es innecesario y lento. En un análisis real, usás una sola conexión para todas las consultas que necesitás en ese script.
 
**Consigna:** Usando una sola conexión a `vetcare.db`, ejecutá estas cuatro consultas y generá un reporte completo de la clínica:
 
1. Total de dueños registrados
2. Distribución de mascotas por especie (cuántos perros, gatos, etc.)
3. La consulta más cara y la más barata
4. Las mascotas que nunca tuvieron una consulta (están en `mascotas` pero no en `consultas`)

```python
import sqlite3
import pandas as pd
 
conexion = sqlite3.connect("vetcare.db")
 
# 1. Total de dueños
total_duenos = pd.read_sql("SELECT COUNT(*) AS total FROM duenos", conexion)
print(f"Dueños registrados: {total_duenos['total'][0]}")
 
# 2. Distribución por especie
df_especies = pd.read_sql("SELECT especie, COUNT(*) AS cantidad FROM mascotas GROUP BY especie", conexion)
print("\nMascotas por especie:")
print(df_especies)
 
# 3. Consulta más cara y más barata
df_extremos = pd.read_sql(
    """
    SELECT motivo, costo, veterinario
    FROM consultas
    WHERE costo = (SELECT MAX(costo) FROM consultas)
       OR costo = (SELECT MIN(costo) FROM consultas)
    ORDER BY costo DESC
    """,
    conexion
)
print("\nConsulta más cara y más barata:")
print(df_extremos)
 
# 4. Mascotas sin consultas
df_sin_consulta = pd.read_sql(
    """
    SELECT mascotas.nombre, mascotas.especie
    FROM mascotas
    LEFT JOIN consultas ON mascotas.id = consultas.mascota_id
    WHERE consultas.id IS NULL
    """,
    conexion
)
print("\nMascotas sin consultas registradas:")
print(df_sin_consulta)
 
conexion.close()
```
