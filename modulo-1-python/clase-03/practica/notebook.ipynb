## Ejercicio 2 — Función con múltiples retornos
 
Un sistema de e-commerce necesita calcular, para cada orden, el subtotal, el descuento aplicado y el total final. Hoy esa lógica está duplicada en tres partes distintas del código.
 
**Consigna:** Define una función `calcular_orden` que reciba el precio unitario, la cantidad y un porcentaje de descuento (por defecto `0`), y devuelva tres valores: el subtotal antes del descuento, el monto del descuento y el total final.
 
Después llamala con estos casos y mostrá los resultados formateados:

```python
# Caso 1: 3 unidades a $12.000, sin descuento
# Caso 2: 5 unidades a $8.500, con 10% de descuento
# Caso 3: 2 unidades a $45.000, con 25% de descuento
 
# Ejemplo de salida esperada para el caso 2:
# Subtotal: $42.500 | Descuento: $4.250 | Total: $38.250
```

````python
def calcular_orden(precio_unitario, cantidad, descuento=0):
    subtotal        = precio_unitario * cantidad
    monto_descuento = subtotal * descuento
    total           = subtotal - monto_descuento
    return subtotal, monto_descuento, total
 
# Caso 1
subtotal, descuento, total = calcular_orden(12000, 3)
print(f"Subtotal: ${subtotal:,.2f} | Descuento: ${descuento:,.2f} | Total: ${total:,.2f}")
 
# Caso 2
subtotal, descuento, total = calcular_orden(8500, 5, descuento=0.10)
print(f"Subtotal: ${subtotal:,.2f} | Descuento: ${descuento:,.2f} | Total: ${total:,.2f}")
 
# Caso 3
subtotal, descuento, total = calcular_orden(45000, 2, descuento=0.25)
print(f"Subtotal: ${subtotal:,.2f} | Descuento: ${descuento:,.2f} | Total: ${total:,.2f}")
````

Este pipeline procesa registros de temperatura de sensores industriales. Ejecutalo exactamente como está.

````python
def convertir_temperatura(registro):
    try:
        celsius = float(registro["valor"])
        return (celsius * 9/5) + 32
    except ValueError:
        return None
 
sensores = [
    {"id": "S-01", "valor": "87.3"},
    {"id": "S-02", "valor": "N/A"},
    {"id": "S-03", "valor": "92.1"},
    {"id": "S-04", "valor": ""},
    {"id": "S-05", "valor": "88.7"},
]
 
for sensor in sensores:
    fahrenheit = convertir_temperatura(sensor)
    if fahrenheit is not None:
        print(f"{sensor['id']}: {fahrenheit:.1f}°F")
    else:
        print(f"{sensor['id']}: sin lectura")
````

## Ejercicio 4 — Pipeline modular
 
Tienes este código que funciona pero está todo junto en un solo bloque. Es difícil de leer, difícil de modificar y si algo falla no sabés dónde.

````python
def limpiar_registro(r):
    if r["producto"] == "":
        raise ValueError("nombre vacío")
 
    precio = float(r["precio"])
    if precio <= 0:
        raise ValueError(f"precio inválido: {r['precio']}")
 
    stock = int(r["stock"])
 
    return {
        "producto":  r["producto"],
        "precio":    precio,
        "stock":     stock,
        "categoria": r["categoria"],
    }
 
def calcular_valor(registro):
    return registro["precio"] * registro["stock"]
 
def agrupar_por_categoria(registros):
    grupos = {}
    for r in registros:
        cat = r["categoria"]
        if cat not in grupos:
            grupos[cat] = []
        grupos[cat].append(r)
    return grupos
 
def procesar_dataset(registros):
    procesados = []
    errores    = []
 
    for r in registros:
        try:
            limpio = limpiar_registro(r)
            limpio["valor"] = calcular_valor(limpio)
            procesados.append(limpio)
        except (ValueError, KeyError) as e:
            errores.append({"producto": r.get("producto", "?"), "error": str(e)})
 
    return procesados, errores
 
 
# --- Programa principal ---
 
registros = [
    {"producto": "Notebook",   "precio": "320000", "stock": "5",   "categoria": "tech"},
    {"producto": "Escritorio", "precio": "85000",  "stock": "N/A", "categoria": "muebles"},
    {"producto": "Mouse",      "precio": "-8500",  "stock": "42",  "categoria": "tech"},
    {"producto": "",           "precio": "45000",  "stock": "8",   "categoria": "muebles"},
    {"producto": "Monitor",    "precio": "210000", "stock": "0",   "categoria": "tech"},
    {"producto": "Auriculares","precio": "32000",  "stock": "11",  "categoria": "tech"},
]
 
procesados, errores = procesar_dataset(registros)
grupos = agrupar_por_categoria(procesados)
 
print(f"Procesados: {len(procesados)} | Errores: {len(errores)}")
 
for categoria, items in grupos.items():
    total = sum(r["valor"] for r in items)
    print(f"\n{categoria.capitalize()} — valor total: ${total:,}")
 
for e in errores:
    print(f"  ✗ {e['producto']}: {e['error']}")
````

**Consigna:** Refactoriza este código separando la lógica en funciones. Cada función debe hacer una sola cosa:
 
- `limpiar_registro(r)` — convierte y valida los tipos, lanza excepciones si algo falla
- `calcular_valor(registro)` — calcula el valor de inventario
- `agrupar_por_categoria(registros)` — agrupa una lista de registros por categoría
- `procesar_dataset(registros)` — coordina todo y devuelve procesados y errores
El resultado final debe ser idéntico al del código original.
